```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

```python
# Generate sample data with a non-linear relationship
x = tf.random.uniform((100, 1), minval = 0, maxval = 5)
```

```python
print(min(x), max(x))
print(x.shape)
```

```python
y = tf.math.sin(x) + x**3 + tf.random.normal((100, 1), mean = 0, stddev = 0.5)
```

```python
print(len(y))
```

```python
plt.figure(figsize = (6, 4))
plt.plot(x, y, '.')
plt.show()
```

```python
print(x.shape)
print(len(y))
```

```python
# Define the neural network model
model = keras.Sequential([
  keras.layers.Dense(100, activation = 'relu', input_shape = (1,)), 
  keras.layers.Dense(100, activation = 'relu'),
  keras.layers.Dense(1)  # Output layer
])
```

```python
# Compile the model with mean squared error loss and optimizer
model.compile(loss = 'mse', optimizer = 'adam')
```

```python
epochs = 2000
batch_size = 32
```

```python
# Train the model on the data
history = model.fit(x, y, epochs = epochs, batch_size = batch_size)
```

```python
# Access training loss history from the 'history' object
training_loss = history.history['loss']
```

```python
# Plot epochs vs. loss
plt.loglog(training_loss)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Loss')
plt.grid(True)
plt.show()
```

```python
y_pred = model.predict(x)
```

```python
plt.figure(figsize = (6, 4))
plt.plot(y, '.')
plt.plot(y_pred, '*')
plt.show()
```
